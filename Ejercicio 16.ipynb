```python
vocales = 0
consonantes = 0

palabra = input("Ingrese una palabra: ")
palabra = palabra.lower()

for letra in palabra:
    if letra == "a" or letra == "e" or letra == "i" or letra == "o" or letra == "u":
        vocales += 1
    else:
        consonantes += 1

print("La palabra tiene", vocales, "vocales y", consonantes, "consonantes.")
```

```python
import numpy as np

# Definir matrices
def definir_matrices():
    print("Definir matrices")
    print("--------------------")

    filas = int(input("Ingrese el número de filas: "))
    columnas = int(input("Ingrese el número de columnas: "))

    matriz = []
    for i in range(filas):
        fila = []
        for j in range(columnas):
            elemento = float(input(f"Ingrese el elemento ({i+1},{j+1}): "))
            fila.append(elemento)
        matriz.append(fila)

    return matriz

# Sumar matrices
def sumar_matrices(matriz1, matriz2):
    if len(matriz1) != len(matriz2) or len(matriz1[0]) != len(matriz2[0]):
        raise ValueError("Las matrices deben tener el mismo tamaño")

    resultado = []
    for i in range(len(matriz1)):
        fila = []
        for j in range(len(matriz1[0])):
            fila.append(matriz1[i][j] + matriz2[i][j])
        resultado.append(fila)

    return resultado

# Restar matrices
def restar_matrices(matriz1, matriz2):
    if len(matriz1) != len(matriz2) or len(matriz1[0]) != len(matriz2[0]):
        raise ValueError("Las matrices deben tener el mismo tamaño")

    resultado = []
    for i in range(len(matriz1)):
        fila = []
        for j in range(len(matriz1[0])):
            fila.append(matriz1[i][j] - matriz2[i][j])
        resultado.append(fila)

    return resultado

# Multiplicar matrices
def multiplicar_matrices(matriz1, matriz2):
    if len(matriz1[0]) != len(matriz2):
        raise ValueError("El número de columnas de la primera matriz debe ser igual al número de filas de la segunda matriz")

    resultado = []
    for i in range(len(matriz1)):
        fila = []
        for j in range(len(matriz2[0])):
            suma = 0
            for k in range(len(matriz1[0])):
                suma += matriz1[i][k] * matriz2[k][j]
            fila.append(suma)
        resultado.append(fila)

    return resultado

# Transponer matriz
def transponer_matriz(matriz):
    resultado = []
    for j in range(len(matriz[0])):
        fila = []
        for i in range(len(matriz)):
            fila.append(matriz[i][j])
        resultado.append(fila)

    return resultado

# Imprimir matriz
def imprimir_matriz(matriz):
    for fila in matriz:
        print(fila)

# Main
def main():
    print("Operaciones con matrices")
    print("---------------------------")

    while True:
        print("1. Definir matriz")
        print("2. Sumar matrices")
        print("3. Restar matrices")
        print("4. Multiplicar matrices")
        print("5. Transponer matriz")
        print("6. Salir")

        opcion = input("Ingrese su opción: ")

        if opcion == "1":
            matriz = definir_matrices()
            print("Matriz definida:")
            imprimir_matriz(matriz)
        elif opcion == "2":
            matriz1 = definir_matrices()
            matriz2 = definir_matrices()
            try:
                resultado = sumar_matrices(matriz1, matriz2)
                print("Resultado de la suma:")
                imprimir_matriz(resultado)
            except ValueError as e:
                print(e)
        elif opcion == "3":
            matriz1 = definir_matrices()
            matriz2 = definir_matrices()
            try:
                resultado = restar_matrices(matriz1, matriz2)
                print("Resultado de la resta:")
                imprimir_matriz(resultado)
            except ValueError as e:
                print(e)
        elif opcion == "4":
            matriz1 = definir_matrices()
            matriz2 = definir_matrices()
            try:
                resultado = multiplicar_matrices(matriz1, matriz2)
                print("Resultado de la multiplicación:")
                imprimir_matriz(resultado)
            except ValueError as e:
                print(e)
        elif opcion == "5":
            matriz = definir_matrices()
            resultado = transponer_matriz(matriz)
            print("Matriz transpuesta:")
            imprimir_matriz(resultado)
        elif opcion == "6":
            print("Hasta luego!")
            break
        else:
            print("Opción inválida. Por favor, intente de nuevo.")

if __name__ == "__main__":
    main()
```

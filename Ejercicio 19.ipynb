```python
import numpy as np
import matplotlib.pyplot as plt

unifome = np.random.uniform(0, 17, 5)
print("distribución uniforme")
print(unifome)

normal = np.random.normal(0, 1, 5)
print("distribución normal")
print(normal)

exponencial = np.random.exponential(1, 5)
print("distribución exponencial")
print(exponencial)



```

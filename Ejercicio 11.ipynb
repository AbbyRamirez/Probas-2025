```python
cadena = input("ingrese una palabra ")
def palindromo(cadena):
  cadena = cadena.lower()
  cadena = ''.join(e for e in cadena if e.isalnum())
  longitud = len(cadena)
  for i in range(longitud // 2):
    if cadena[i] != cadena[longitud - 1 - i]:
      return False
  return True
if palindromo(cadena):
  print("es palindromo")
else:
  print("no es palindromo")
```

```python
def convertir_longitud():
    print("Conversor de longitud")
    print("------------------------")
    longitud = float(input("Ingrese la longitud a convertir: "))
    unidad = input("Ingrese la unidad de la longitud (m, cm, mm, km, in, ft, yd): ")

    if unidad == "m":
        print(f"{longitud} metros son iguales a:")
        print(f"{longitud * 100} centímetros")
        print(f"{longitud * 1000} milímetros")
        print(f"{longitud / 1000} kilómetros")
        print(f"{longitud * 39.37} pulgadas")
        print(f"{longitud * 3.2808} pies")
        print(f"{longitud * 1.0936} yardas")
    elif unidad == "cm":
        print(f"{longitud} centímetros son iguales a:")
        print(f"{longitud / 100} metros")
        print(f"{longitud * 10} milímetros")
        print(f"{longitud / 100000} kilómetros")
        print(f"{longitud * 0.3937} pulgadas")
        print(f"{longitud * 0.032808} pies")
        print(f"{longitud * 0.010936} yardas")
    elif unidad == "mm":
        print(f"{longitud} milímetros son iguales a:")
        print(f"{longitud / 1000} metros")
        print(f"{longitud / 10} centímetros")
        print(f"{longitud / 1000000} kilómetros")
        print(f"{longitud * 0.03937} pulgadas")
        print(f"{longitud * 0.0032808} pies")
        print(f"{longitud * 0.0010936} yardas")
    elif unidad == "km":
        print(f"{longitud} kilómetros son iguales a:")
        print(f"{longitud * 1000} metros")
        print(f"{longitud * 100000} centímetros")
        print(f"{longitud * 1000000} milímetros")
        print(f"{longitud * 39370} pulgadas")
        print(f"{longitud * 3280.8} pies")
        print(f"{longitud * 1093.6} yardas")
    elif unidad == "in":
        print(f"{longitud} pulgadas son iguales a:")
        print(f"{longitud / 39.37} metros")
        print(f"{longitud / 0.3937} centímetros")
        print(f"{longitud / 0.03937} milímetros")
        print(f"{longitud / 39370} kilómetros")
        print(f"{longitud * 0.083333} pies")
        print(f"{longitud * 0.027778} yardas")
    elif unidad == "ft":
        print(f"{longitud} pies son iguales a:")
        print(f"{longitud / 3.2808} metros")
        print(f"{longitud / 0.032808} centímetros")
        print(f"{longitud / 0.0032808} milímetros")
        print(f"{longitud / 3280.8} kilómetros")
        print(f"{longitud * 12} pulgadas")
        print(f"{longitud * 0.33333} yardas")
    elif unidad == "yd":
        print(f"{longitud} yardas son iguales a:")
        print(f"{longitud / 1.0936} metros")
        print(f"{longitud / 0.010936} centímetros")
        print(f"{longitud / 0.0010936} milímetros")
        print(f"{longitud / 1093.6} kilómetros")
        print(f"{longitud * 36} pulgadas")
        print(f"{longitud * 3} pies")

def convertir_masa():
    print("Conversor de masa")
    print("--------------------")
    masa = float(input("Ingrese la masa a convertir: "))
    unidad = input("Ingrese la unidad de la masa (kg, g, mg, t, lb, oz): ")

    if unidad == "kg":
        print(f"{masa} kilogramos son iguales a:")
        print(f"{masa * 1000} gramos")
        print(f"{masa * 1000000} miligramos")
        print(f"{masa / 1000} toneladas")
        print(f"{masa * 2.20462} libras")
```

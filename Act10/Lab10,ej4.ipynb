```python
import math

# Función pa
def calcular_estadisticas(*args):

    promedio = lambda nums: sum(nums) / len(nums)

    def mediana(nums):
        nums = sorted(nums)
        n = len(nums)
        if n % 2 == 0:
            return (nums[n // 2 - 1] + nums[n // 2]) / 2
        else:
            return nums[n // 2]

    def desviacion_estandar(nums):
        prom = promedio(nums)
        return math.sqrt(sum((x - prom) ** 2 for x in nums) / len(nums))


    avg = promedio(args)
    median = mediana(args)
    std_dev = desviacion_estandar(args)


    return avg, median, std_dev


numeros = list(map(float, input("Introduce una lista de números separados por espacio: ").split()))

promedio, mediana, desviacion = calcular_estadisticas(*numeros)

print(f"Promedio: {promedio}")
print(f"Mediana: {mediana}")
print(f"Desviación Estándar: {desviacion}")
```

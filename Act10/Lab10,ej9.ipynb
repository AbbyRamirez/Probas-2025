```python

class Producto:
    def __init__(self, nombre, precio, cantidad):
        self.nombre = nombre
        self.precio = precio
        self.cantidad = cantidad

    def __str__(self):
        return f"{self.nombre} - ${self.precio} - Cantidad: {self.cantidad}"

    def actualizar_stock(self, cantidad_vendida):
        if cantidad_vendida <= self.cantidad:
            self.cantidad -= cantidad_vendida
        else:
            print("No hay suficiente stock.")


from modulo_producto import Producto

def agregar_producto(inventario, nombre, precio, cantidad):
    producto = Producto(nombre, precio, cantidad)
    inventario.append(producto)

def mostrar_inventario(inventario):
    for producto in inventario:
        print(producto)

def calcular_total(inventario, compras):
    total = 0
    for producto, cantidad in compras.items():
        for p in inventario:
            if p.nombre == producto:
                total += p.precio * cantidad
                p.actualizar_stock(cantidad)
                break
    return total

def mostrar_menu():
    print("Menú:")
    print("1. Agregar Producto")
    print("2. Mostrar Inventario")
    print("3. Realizar Compra")
    print("4. Salir")

def obtener_opcion():
    try:
        opcion = int(input("Elija una opción: "))
        return opcion
    except ValueError:
        print("Por favor ingrese un número válido.")
        return obtener_opcion()

from modulo_inventario import agregar_producto, mostrar_inventario
from modulo_compras import calcular_total
from modulo_usuario import mostrar_menu, obtener_opcion

class Tienda:
    def __init__(self):
        self.inventario = []
        self.compras = {}

    def agregar_producto(self):
        nombre = input("Ingrese el nombre del producto: ")
        precio = float(input("Ingrese el precio del producto: "))
        cantidad = int(input("Ingrese la cantidad disponible: "))
        agregar_producto(self.inventario, nombre, precio, cantidad)
        print(f"Producto {nombre} agregado al inventario.")

    def realizar_compra(self):
        print("Productos disponibles para compra:")
        mostrar_inventario(self.inventario)
        producto = input("Ingrese el nombre del producto a comprar: ")
        cantidad = int(input(f"¿Cuántos de {producto} desea comprar?: "))
        if producto in self.compras:
            self.compras[producto] += cantidad
        else:
            self.compras[producto] = cantidad
        total = calcular_total(self.inventario, self.compras)
        print(f"Total de la compra: ${total}")

    def mostrar_inventario(self):
        print("Inventario actual:")
        mostrar_inventario(self.inventario)

    def menu(self):
        while True:
            mostrar_menu()
            opcion = obtener_opcion()

            if opcion == 1:
                self.agregar_producto()
            elif opcion == 2:
                self.mostrar_inventario()
            elif opcion == 3:
                self.realizar_compra()
            elif opcion == 4:
                print("Saliendo del programa.")
                break
            else:
                print("Opción no válida, por favor intente de nuevo.")

from clase_tienda import Tienda

def main():
    tienda = Tienda()
    tienda.menu()

if __name__ == "__main__":
    main()
```

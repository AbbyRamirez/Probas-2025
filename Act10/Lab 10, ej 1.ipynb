```python
def resumen (cadena):
  cadena = cadena.lower()
  cadena = ''.join(e for e in cadena if e.isalnum())
  print(len(cadena))

  st = set()

  for letra in cadena:
    letter = cadena.count(letra)
    if letter == 1:
      st.add(letra)

  print(st)

  dc2 = dict()

  for letra in cadena:
    llave = cadena.count(letra)
    dc2.update({letra:llave})

  print(dc2)

  mas = 1
  rep = ''
  for letra in cadena:
    if dc2[letra] > mas:
      mas = dc2[letra]
      rep = letra

  print(rep)

cadena = str(input("diga una palabra:"))
resumen(cadena)
```

```python

class Vehiculo:
    def __init__(self, marca, modelo, ano, precio):
        self.marca = marca
        self.modelo = modelo
        self.ano = ano
        self.precio = precio

    def mostrar_info(self):
        return f"Marca: {self.marca}\nModelo: {self.modelo}\nAño: {self.ano}\nPrecio: {self.precio}"


class Automovil(Vehiculo):
    def __init__(self, marca, modelo, ano, precio, num_puertas):
        super().__init__(marca, modelo, ano, precio)
        self.num_puertas = num_puertas

    def mostrar_info(self):
        info = super().mostrar_info()
        return f"{info}\nNúmero de puertas: {self.num_puertas}"


class Motocicleta(Vehiculo):
    def __init__(self, marca, modelo, ano, precio, cilindrada):
        super().__init__(marca, modelo, ano, precio)
        self.cilindrada = cilindrada

    def mostrar_info(self):
        info = super().mostrar_info()
        return f"{info}\nCilindrada: {self.cilindrada}"


auto = Automovil("Toyota", "Corolla", 2022, 20000, 4)
moto = Motocicleta("Yamaha", "MT-07", 2023, 8000, 689)

print(auto.mostrar_info())
print("\n" + "-"*30 + "\n")
print(moto.mostrar_info())
```

```python
tru = True
while tru == True:
  print("1: agregar producto")
  print("2: eliminar producto")
  print("3: buscar producto")
  print("4: ordenar productos")
  choice = int(input("seleccione el numero correspondiente a la accion que desea hacer"))

  lista = []

  if choice == 1:
    a = input("ingrese nombre del producto")
    a = {"Nombre":a,
                 "categoria":input("ingrese categoria"),
                 "precio": input("ingrese precio"),
                 "cantidad": input("ingrese cantidad")}
    lista.append(a)
    print("producto agregado")

  elif choice == 2:
   busqueda = input("ingrese el producto a eliminar")
   for i in lista:
      if i["Nombre"]== busqueda:
         lista.remove(i)
         print("producto eliminado")

  elif choice == 3:
    busqueda = input("ingrese el producto que busca")
    for i in lista:
      if i["Nombre"] == busqueda:
       print(i)

  elif choice == 4:
   lista.sort()
   print(lista)

tru = True
```

```python
class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.siguiente = None

class ListaEnlazada:
    def __init__(self):
        self.cabeza = None

    def agregar(self, valor):
        nodo = Nodo(valor)
        if not self.cabeza:
            self.cabeza = nodo
        else:
            actual = self.cabeza
            while actual.siguiente:
                actual = actual.siguiente
            actual.siguiente = nodo

    def imprimir(self):
        actual = self.cabeza
        while actual:
            print(actual.valor, end=" ")
            actual = actual.siguiente
        print()

class Pila:
    def __init__(self):
        self.elementos = []

    def push(self, valor):
        self.elementos.append(valor)

    def pop(self):
        if self.vacio():
            return None
        return self.elementos.pop()

    def vacio(self):
        return len(self.elementos) == 0

    def imprimir(self):
        print(self.elementos)
class Cola:
    def __init__(self):
        self.elementos = []

    def enqueue(self, valor):
        self.elementos.append(valor)

    def dequeue(self):
        if self.vacio():
            return None
        return self.elementos.pop(0)

    def vacio(self):
        return len(self.elementos) == 0

    def imprimir(self):
        print(self.elementos)

lista_enlazada = ListaEnlazada()
lista_enlazada.agregar(9)
lista_enlazada.agregar(0)
lista_enlazada.agregar(4)
lista_enlazada.imprimir()

pila = Pila()
pila.push(7)
pila.push(4)
pila.push(1)
pila.imprimir()

cola = Cola()
cola.enqueue(7)
cola.enqueue(8)
cola.enqueue(4)
cola.imprimir()
print(cola.dequeue())
cola.imprimir()
```

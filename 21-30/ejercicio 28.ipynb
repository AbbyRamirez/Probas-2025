```python
def crear_cuenta(numero_cuenta, saldo_inicial):
    cuenta = {
        "numero_cuenta": numero_cuenta,
        "saldo": saldo_inicial
    }
    return cuenta

def depositar(cuenta, monto):
    cuenta["saldo"] += monto
    print(f"Depósito de ${monto} realizado con éxito. Saldo actual: ${cuenta['saldo']}")

def retirar(cuenta, monto):
    if cuenta["saldo"] >= monto:
        cuenta["saldo"] -= monto
        print(f"Retiro de ${monto} realizado con éxito. Saldo actual: ${cuenta['saldo']}")
    else:
        print("Fondos insuficientes para realizar el retiro.")

def consultar_saldo(cuenta):
    print(f"Saldo actual: ${cuenta['saldo']}")

def main():
    print("Simulador de cuenta bancaria")
    print("---------------------------------")

    numero_cuenta = input("Ingrese el número de cuenta: ")
    saldo_inicial = float(input("Ingrese el saldo inicial: "))

    cuenta = crear_cuenta(numero_cuenta, saldo_inicial)

    while True:
        print("Opciones:")
        print("1. Depositar")
        print("2. Retirar")
        print("3. Consultar saldo")
        print("4. Salir")

        opcion = input("Ingrese su opción: ")

        if opcion == "1":
            monto = float(input("Ingrese el monto a depositar: "))
            depositar(cuenta, monto)
        elif opcion == "2":
            monto = float(input("Ingrese el monto a retirar: "))
            retirar(cuenta, monto)
        elif opcion == "3":
            consultar_saldo(cuenta)
        elif opcion == "4":
            print("Hasta luego!")
            break
        else:
            print("Opción inválida. Por favor, intente de nuevo.")

if __name__ == "__main__":
    main()
```

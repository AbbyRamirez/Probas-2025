```python
def suma_serie():
    print("Suma de una serie numérica")
    print("--------------------------------")

    n = int(input("Ingrese el número de términos: "))
    serie = []

    for i in range(n):
        termino = float(input(f"Ingrese el término {i+1}: "))
        serie.append(termino)

    suma = sum(serie)

    print(f"La suma de la serie es: {suma}")

def suma_serie_aritmetica():
    print("Suma de una serie aritmética")
    print("---------------------------------")

    n = int(input("Ingrese el número de términos: "))
    a1 = float(input("Ingrese el primer término: "))
    d = float(input("Ingrese la diferencia común: "))

    suma = n * (2 * a1 + (n - 1) * d) / 2

    print(f"La suma de la serie aritmética es: {suma}")

def suma_serie_geometrica():
    print("Suma de una serie geométrica")
    print("---------------------------------")

    n = int(input("Ingrese el número de términos: "))
    a1 = float(input("Ingrese el primer término: "))
    r = float(input("Ingrese la razón común: "))

    if r == 1:
        suma = n * a1
    else:
        suma = a1 * (1 - r ** n) / (1 - r)

    print(f"La suma de la serie geométrica es: {suma}")

def main():
    print("Suma de series numéricas")
    print("----------------------------")

    while True:
        print("1. Suma de una serie numérica")
        print("2. Suma de una serie aritmética")
        print("3. Suma de una serie geométrica")
        print("4. Salir")

        opcion = input("Ingrese su opción: ")

        if opcion == "1":
            suma_serie()
        elif opcion == "2":
            suma_serie_aritmetica()
        elif opcion == "3":
            suma_serie_geometrica()
        elif opcion == "4":
            print("Hasta luego!")
            break
        else:
            print("Opción inválida. Por favor, intente de nuevo.")

if __name__ == "__main__":
    main()
```

```python
limite = int(input("seleccione limite "))
par = list()
impar = list()
impresion = int()

for num in range(1,limite+1):
  if num % 2 == 0:
    par.append(num)
  else:
    impar.append(num)

impresion = max(len(par),len(impar))

for item in range(impresion):
  try:
     print(par[item],impar[item])
  except:


    try:
      v1 = par[item]
    except:
      v1 = 0
    try:
      v2 = impar[item]
    except:
      v2 = 0
    finally:
      print(v1,v2)







```
